`Summing the elements of a sequence using linear recursion.` 
- Uses `O(n)` since it has n+1 function calls totally.

```python
def linear_sum(S, n):
    if n == 0:
        return 0
    else:
        return linear_sum(S, n - 1) + S[n - 1]


dts = [4, 7, 1, 9, 2]
sall = linear_sum(dts, len(dts))
print(f"Sum of all elements of the container is : {sall} ")
```

`Reversing a sequence with recursion`
- total number of floor(n / 2) function calls are made. total runtime is `O(n)`

```python
import random

def reverse(S, start, stop):
    if start < stop - 1 :
        S[start] , S[stop -1] = S[stop - 1], S[start]
        reverse(S, start + 1, stop - 1)


test_data = [random.randint(1, 50) for i in range(33)]
reverse(test_data, 0,32 )
print(test_data)
```

`Problem of raising a number x to an arbitrary nonnegative integer, n.`


```python
def power(x, n):
    if n == 0:
        return 1
    else:
        return power(x, n -1 ) * x

p = power(2, 6)
print(p)
```
